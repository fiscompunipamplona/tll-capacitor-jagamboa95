Considere el siguiente modelo simple de un capacitor elctronico,consiste en dos placas mealicas planas en una caja de metal cuadrada:

Por simplificar vamos a modelar el sistema en dos dimensiones. Utilizando cualquiera de los metodos estudiados, escriba un programa para calcular el potncial electrostatico en la caja en una malla de $100x100$ puntos, donde los muros de la caja tienen voltaje cero y las dos placas ( las cuales tienen grosor despresiable) tienen un voltaje de $+V$ o $-V$ como se muestra.El programa debe calcular el valor del potencial en cada punto de la malla con una presici√≥n de  $10^{-6}$ voltios y haga una grafica de la densidad.

```python
from numpy import empty, zeros ,max
from pylab import imshow, gray, show,hot,cool,inferno
```

## metodo d Gauss-seidel con grosor de la placa de 1mm

```python
M=100
V=1.0 
target=1e-3


phi=zeros([M+1,M+1],float)
phi[0,:]=0
rho=zeros([M+1,M+1],float) 
rho[20:80,80]= -V
rho[20:80,20]= V

delta=1.0
omega=0.9

while delta>target:
    delta=0
    for i in range(1,M):
        for j in range(1,M):
            difference =(phi[i+1,j]+phi[i-1,j]\
                +phi[i,j+1]+phi[i,j-1]+rho[i,j])/4-phi[i,j]
            phi[i,j]=phi[i,j]+(1+omega)*difference
            
            if difference>delta: delta=difference

imshow(phi)
inferno()
show()
```
